```python
# Import relevant libraries
import numpy as np
from matplotlib.pylab import plt
from matplotlib.animation import FuncAnimation
```

```python
# Write an animation of two sinusoidal waves with different wavelengths and frequencies, and different phase-speeds
def animate_sinusoids(wavelength1, frequency1, phase_speed1, wavelength2, frequency2, phase_speed2, duration, fps=30, save_gif=True, filename='wave_animation.gif'):

    # Calculate the time array
    t = np.linspace(0, duration, int(fps * duration))
    
    # Calculate the spatial array (cover both wavelengths)
    x_min = 0
    x_max = max(wavelength1, wavelength2)
    x = np.linspace(x_min, x_max, 1000)
    
    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Set the axis limits
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(-3, 3)
    ax.set_xlabel('Position')
    ax.set_ylabel('Amplitude')
    ax.set_title('Sinusoidal Waves with Different Frequencies, Wavelengths, and Phase Speeds')
    
    # Initialize the lines for the two waves and their sum
    line1, = ax.plot([], [], 'b-', linewidth=2, label=f'Wave 1: λ={wavelength1}, f={frequency1}, v={phase_speed1}')
    line2, = ax.plot([], [], 'r-', linewidth=2, label=f'Wave 2: λ={wavelength2}, f={frequency2}, v={phase_speed2}')
    line_sum, = ax.plot([], [], 'g-', linewidth=2, label='Sum of Waves')
    
    # Add a legend
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Animation function
    def update(frame):
        y1 = np.sin(2 * np.pi * (x / wavelength1 - frequency1 * t[frame]))
        y2 = np.sin(2 * np.pi * (x / wavelength2 - frequency2 * t[frame]))
        y_sum = y1 + y2
        line1.set_data(x, y1)
        line2.set_data(x, y2)
        line_sum.set_data(x, y_sum)
        return line1, line2, line_sum
    
    # Create the animation
    ani = FuncAnimation(fig, update, frames=len(t), interval=1000/fps, blit=True, repeat=True)
    
    if save_gif:
        # Save as GIF
        print(f"Saving animation as {filename}...")
        ani.save(filename, writer='pillow', fps=fps)
        print(f"Animation saved successfully as {filename}")
    else:
        plt.show()
    
    return ani

wavelength1 = 2.0  # Wavelength of the first wave
wavelength2 = 2.0  # Wavelength of the second wave

frequency1 = 0.2   # Frequency of the first wave
frequency2 = 0.1   # Frequency of the second wave

phase_speed1 = frequency1 * wavelength1  # Phase speed of the first wave
phase_speed2 = frequency2 * wavelength2  # Phase speed of the second wave

# Example usage
animate_sinusoids(
    wavelength1=wavelength1, frequency1=frequency1, phase_speed1=phase_speed1,
    wavelength2=wavelength2, frequency2=frequency2, phase_speed2=phase_speed2,
    duration=20, save_gif=True, filename='sinusoidal_waves.gif'
)
```
