# Solving for the Boundaries of Mihalas & Toomre 1981 & Souffrin 1966

```python
# Import the relevant libraries
import sympy as sp
from sympy import sqrt, sin, cos, Rational
from sympy import Pow

# Define the coefficients
g = sp.Symbol('g')
cs = sp.Symbol('c_s')
N = sp.Symbol('N')
H = sp.Symbol('H')
omega = sp.Symbol('omega')
wac = sp.Symbol('w_{ac}')
gamma = sp.Symbol('gamma')
tau = sp.Symbol('tau')
kh = sp.Symbol('k_h')

# Define the equation for Mihalas and Toomre 1981 & Souffrin 1966
eq_mt81 = (pow(N, 2)*pow(kh, 2)/pow(omega, 2) + pow(omega, 2)/pow(cs, 2) -1/(4*pow(H, 2)) - pow(kh, 2))
eq_sf66 = (pow(omega, 2) - pow(wac, 2))/pow(cs, 2) - (pow(omega, 2) - pow(N, 2))*pow(kh, 2)/pow(omega, 2)

# Solve the dispersion relation for omega
roots_mt81 = sp.solve(eq_mt81, omega)
roots_sf66 = sp.solve(eq_sf66, omega)

# Print the results in python code
print('Solutions for the Mihalas and Toomre 1981 dispersion relation')
print('---------------------------------------------------------------------------------')
print(sp.python(roots_mt81[0]))
print(sp.python(roots_mt81[1]))
print(sp.python(roots_mt81[2]))
print(sp.python(roots_mt81[3]))
print('\nSolutions for the Souffrin 1966 dispersion relation')
print('-------------------------------------------------------------------')
print(sp.python(roots_sf66[0]))
print(sp.python(roots_sf66[1]))
print(sp.python(roots_sf66[2]))
print(sp.python(roots_sf66[3]))
roots_mt81[1]
```

# Solving for the Boundaries of Mihalas & Toomre 1982 

```python
# # Define the variables for the Mihalas & Toomre 1982 dispersion relation
h0 = N**2*kh**2*omega**(-2) - kh**2 + omega**2*cs**(-2) - 1*(4*H**2)**(-1)
eta = omega*tau
a0 = h0 + (1 + eta**2)**(-1)*((gamma - 1)*omega**2*cs**(-2) - (N**2*kh**2)*omega**(-2))
b0 = eta*(1 + eta**2)**(-1)*((gamma - 1)*omega**2*cs**(-2) - (N**2*kh**2)*omega**(-2))

# Print the equation for kz^2
eq = 0.5*(a0 + (a0**2 + b0**2)**(0.5))
print('kz^2 = \n', sp.python(eq))
```

# Solving for the Boundaries of Bunte & Bogdan 1993

```python
# Define the variables for the Bunte & Bogdan 1993 dispersion relation
a = sp.Symbol('a')
epsilon = sp.Symbol('epsilon')
a02 = sp.Symbol('a02')
a2 = sp.Symbol('a2')
gamma_hat = (1 - 1j*omega*tau*gamma)*(1 - 1j*omega*tau)**(-1)
c2 = gamma*cs**2
c2_hat = gamma_hat*c2

p1 = (cs**2 + a**2)*omega**2 - cs**2*a**2*kh**2
p_gamma = (cs**2 + gamma*a**2)*omega**2 - cs**2*a**2*kh**2
g1 = omega**4 + g**2*(cs**2/(g*H) - 1)*kh**2
g_gamma = gamma*omega**4 + gamma*g**2*(cs**2/(gamma*g*H) - 1)*kh**2
hr = -(1/(4*H**2)) - kh**2 + (p_gamma*g_gamma + (omega*tau*gamma)**2*p1*g1) / (p_gamma**2 + (omega*tau*gamma)**2*p1**2)
hi = (omega*tau*gamma)*(p1*g_gamma - p_gamma*g1)/(p_gamma**2 + (omega*tau*gamma)**2*p1**2)

# Find common denominator
hr_common = sp.simplify(hr)

# Take the numerator
hr_num = sp.numer(hr_common)

# Expand the numerator
hr_num_exp = sp.expand(hr_num)

# Group the terms by omega
hr_num_group = sp.collect(hr_num_exp, omega)
coeff0 = hr_num_group.coeff(omega, 0)
coeff1 = hr_num_group.coeff(omega, 1)
coeff2 = hr_num_group.coeff(omega, 2)
coeff3 = hr_num_group.coeff(omega, 3)
coeff4 = hr_num_group.coeff(omega, 4)
coeff5 = hr_num_group.coeff(omega, 5)
coeff6 = hr_num_group.coeff(omega, 6)
coeff7 = hr_num_group.coeff(omega, 7)
coeff8 = hr_num_group.coeff(omega, 8)

# Print the results in python code
print(sp.python(coeff0))
print(sp.python(coeff1))
print(sp.python(coeff2))
print(sp.python(coeff3))
print(sp.python(coeff4))
print(sp.python(coeff5))
print(sp.python(coeff6))
print(sp.python(coeff7))
print(sp.python(coeff8))
```

# Solving for the Boundaries of Newington & Cally 2009

### Solving for $k_z$

```python
# Make symbols
phi = sp.Symbol('phi')
theta = sp.Symbol('theta')
a = sp.Symbol('a')
ax = sp.Symbol('ax')#a*sp.sin(theta)*sp.cos(phi)
ay = sp.Symbol('ay')#a*sp.sin(theta)*sp.sin(phi)
az = sp.Symbol('az')#a*sp.cos(theta)
kz = sp.Symbol('kz')
kh = sp.Symbol('kh')
k = sp.sqrt((kh**2 + kz**2))
k_par = kh*sp.sin(theta)*sp.cos(phi) + kz*sp.cos(theta)

# Define dispersion relation (Magnetoacoustic Stratification + Alfven Wave Coupling)
D = omega**2*wac**2*ay**2*kh**2 + (omega**2 - a**2*k_par**2)*(omega**4 - (a**2 + cs**2)*omega**2*k**2 + a**2*cs**2*k**2*k_par**2 + cs**2*N**2*kh**2 - (omega**2 - az**2*k**2)*wac**2)

# # Magnetoacoustic stratification
# D = omega**4 - (a**2 + cs**2)*omega**2*k**2 + a**2*cs**2*k**2*k_par**2 + cs**2*N**2*kh**2 - (omega**2 - az**2*k**2)*wac**2

# # Magnetoacoustic NO stratification
# D = omega**4 - (a**2 + cs**2)*omega**2*k**2 + a**2*cs**2*k**2*k_par**2

# # Alfven Waves
# D = omega**2 - a**2*k_par**2

# Group by factors of kz
D_group_kz = sp.collect(sp.expand(D),  kz)

# Extract coefficients of kz
coeff0 = D_group_kz.coeff(kz, 0)
coeff1 = D_group_kz.coeff(kz, 1)
coeff2 = D_group_kz.coeff(kz, 2)
coeff3 = D_group_kz.coeff(kz, 3)
coeff4 = D_group_kz.coeff(kz, 4)
coeff5 = D_group_kz.coeff(kz, 5)
coeff6 = D_group_kz.coeff(kz, 6)

# Print the results in python code
print(sp.python(coeff0))
print(sp.python(coeff1))
print(sp.python(coeff2))
print(sp.python(coeff3))
print(sp.python(coeff4))
print(sp.python(coeff5))
print(sp.python(coeff6))
```

### Solving for $\omega$

```python
# Group the oth-order coefficent from the dispersion relation by omega
coeff0_group = sp.collect(sp.expand(coeff0), omega)

# Extract coefficients of omega
omega_coeff0 = coeff0_group.coeff(omega, 0)
omega_coeff1 = coeff0_group.coeff(omega, 1)
omega_coeff2 = coeff0_group.coeff(omega, 2)
omega_coeff3 = coeff0_group.coeff(omega, 3)
omega_coeff4 = coeff0_group.coeff(omega, 4)
omega_coeff5 = coeff0_group.coeff(omega, 5)
omega_coeff6 = coeff0_group.coeff(omega, 6)

# Print the results in python code
print(sp.python(omega_coeff0))
print(sp.python(omega_coeff1))
print(sp.python(omega_coeff2))
print(sp.python(omega_coeff3))
print(sp.python(omega_coeff4))
print(sp.python(omega_coeff5))
print(sp.python(omega_coeff6))
```

### Solving for Alfven Boundary

```python
a = sp.Symbol('a')
ax = sp.Symbol('ax')
ay = sp.Symbol('ay')
az = sp.Symbol('az')

# Use a u-substitution (u = omega**2)
u = sp.Symbol('u')  

eq = u**3*kh**2 - u**2*((a**2 + ax**2 +cs**2)*kh**2 + wac**2) + u*((2*cs**2 + a**2)*ax**2*kh**2 + wac**2*a**2 + N**2*cs**2) - (N**2*cs**2 + wac**2*az**2 + cs**2*ax**2*kh**2)*ax**2

# Solve for u
roots = sp.solve(eq, u)

# Print the roots in python code
print(sp.python(roots[0]))
print(sp.python(roots[1]))
print(sp.python(roots[2]))
```
